```python
import pandas as pd

# Load the CSV file into a DataFrame
electric_vehicles = pd.read_csv('sub_elect_vehicles.csv')

# Clean the column names: remove underscores and strip extra spaces
#electric_vehicles.columns = electric_vehicles.columns.str.replace('_', '').str.strip()

# Save the cleaned DataFrame back to a new CSV file
electric_vehicles.to_csv('sub_elect_vehicles.csv', index=False)

# Print cleaned columns to verify
print(electric_vehicles.columns)
```

```python
#convert the electric range from miles to kilometers:
sub_elect_vehicles['Electric Range']=sub_elect_vehicles['Electric Range']*1.60934
```

```python
# Total number of all electric vehicles registered by vehicle type countplot
plt.figure(figsize=(5, 2))
sns.countplot(x='Electric Vehicle Type', data=sub_elect_vehicles, color='purple')
plt.title('Total Number of All Electric Vehicles Registered by Vehicle Type', color='red')
plt.xlabel('Electric Vehicle Type', color='green')
plt.show()
```

```python
#market evolvement in the last 10 years
sub_elect_growth=sub_elect_vehicles.groupby(['Model Year'])['DOL Vehicle ID'].nunique()
sub_elect_growth
```

```python
#market evolvement in the last 10yrs bar chart
plt.figure(figsize=(6,3))
sub_elect_growth.plot(kind='bar')
plt.title('Market Growth for Registered Vehicles (2014-2023)', color='red')
plt.xlabel('Years', color='green')
plt.ylabel('Vehicles Number', color='green')
plt.grid(True)
plt.legend()
```

```python
# bev market growth line chart:
plt.figure(figsize=(6,3))
sub_elect_growth.plot(kind='line')
plt.title('BEV Market Growth For Registered Vehicles (2014-2023)',color='purple')
plt.xlabel('Years')
plt.ylabel('Vehicles Number')
plt.grid(True)
plt.legend()
```

```python
# evolvment of the phev market in th 10 yrs:
phev_mt_growth=phev_veh.groupby(['Model Year'])['DOL Vehicle ID'].nunique()
phev_mt_growth
```

```python
#PHEV Market Growth For Registered Vehicles (2014-2023)bar chart:
plt.figure(figsize=(6,3))
phev_mt_growth.plot(kind='bar')
plt.title('PHEV Market Growth For Registered Vehicles (2014-2023)',color='purple')
plt.xlabel('Years')
plt.ylabel('Vehicles Number')
plt.legend()
```

```python
# PHEV Market Growth For Registered Vehicles (2014-2023) line chart:
plt.figure(figsize=(6,3))
phev_mt_growth.plot(kind='line')
plt.title('PHEV Market Growth For Registered Vehicles (2014-2023)',color='purple')
plt.xlabel('Years')
plt.ylabel('Vehicles Number')
plt.grid(True)
plt.legend()
```

```python
#grouped lines chart for the market growth in 10 yrs:
# Create a DataFrame from the aggregated data
df = pd.DataFrame({
'Year': sub_elect_growth.index,
'All Electric': sub_elect_growth.values, # Add sub_elect_vehicles data
'Battery Electric': sub_elect_growth.values,
'Plug-in Hybrid': phev_mt_growth.values
})
# Set 'Year' as the index
df.set_index('Year', inplace=True)
# Plot the grouped line chart
plt.figure(figsize=(12, 6))
df.plot(kind='line', marker='o')
plt.title(' Detailed Market Growth For Registered Vehicles (2014-2023) - Grouped Line Chart',color='purple')
plt.xlabel('Years', color='purple')
plt.ylabel('Vehicles Number', color='purple')
plt.grid(True)
plt.legend()
plt.show()
```
